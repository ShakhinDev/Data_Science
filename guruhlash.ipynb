# Guruhlash

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv', usecols=['country', 'area','pop2021'] ,index_col='country')
df.head()
```

```python
bins = [0,1000, 10000, 30000, 50000, 100000, 300000, 1500000] # bins degan oraliq yaratib oldik
```

```python
population = df.pop2021
```

```python
guruh = pd.cut(population , bins)
guruh
```

```python
guruh.value_counts()
```

```python
df['guruh']= guruh
df.head()
```
