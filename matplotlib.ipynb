# Ustunli diogrammalar sns.barplot va Heatmap bilan tanishamiz

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0,parse_dates=[0], usecols=[0,1])
df.head() # parse_dates bu bizdagi indexni dataindex korinishiga otkazadi , bu bilan biz bemalol oyni yilni kunni ajratib olishimiz mumkin
```

```python
df['month'] = [i.month for i in df.index] # df.index dagi har bitta oyni olib month degan ustunga saqlash
df.head()
```

```python
df['year'] = [i.year for i in df.index]
```

```python
df1=df.groupby(['month', 'year']).max()
```

```python
df1
```

```python
df1.unstack(level=1)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df1.unstack(level=1), cmap="YlGnBu", linewidths=1, square=True)
xticks_labels=['Jan','Feb', 'Mar', 'Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
plt.xticks(np.arange(12), labels=xticks_labels)
plt.show()
```
